```python
import tensorflow as tf
import numpy as np
```

```python
def add_layer(inputs, in_size, out_size, activ_func = None):
    w = tf.Variable(tf.random_normal([in_size, out_size]))
    b = tf.Variable(tf.zeros([1, out_size]) + 0.1)
    f = tf.matmul(inputs, w) + b
    if activ_func is None:
        outputs = f
    else:
        outputs = activ_func(f)
    return outputs
```

```python
#create test data
x_data = np.linspace(-1, 1, 300, dtype = np.float32) #generate a column of data
noise = np.random.normal(0, 0.05, x_data.shape)  #generate normal data and it's loc=0
y_data = np.square(x_data) - 0.5 + noise
```

```python
x_data = x_data.reshape((300,1))
x_data.shape
y_data = y_data.reshape((300,1))
```

```python
#give tensorflow input placeholder
xs = tf.placeholder(dtype = tf.float32, shape = [None,1])
ys = tf.placeholder(dtype = tf.float32, shape = [None,1])
l1 = add_layer(x_data, 1, 10, activ_func=tf.nn.relu)
prediction = add_layer(l1, 10, 1, activ_func=None)

loss = tf.reduce_mean(tf.reduce_sum(tf.square(y_data - prediction),reduction_indices=[1])) #,reduction_indices=[1])
train = tf.train.GradientDescentOptimizer(0.01).minimize(loss)

init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    for i in range(10000):
        sess.run(train, feed_dict = {xs: x_data, ys: y_data})
        if(i % 10 == 0):
            print(sess.run(loss, feed_dict={xs:x_data,ys:y_data}))
```
